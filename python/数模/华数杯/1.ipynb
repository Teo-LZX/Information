```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

#读取厚度
thickness = pd.read_excel(io='./C题数据.xlsx', usecols=['厚度mm'])
thickness_before = np.array(thickness)[:-1:2]   #插层之前的厚度
thickness_after = np.array(thickness)[1::2]     #插层之后的厚度
#print(thickness_before, len(thickness_before))          
#print(thickness_after, len(thickness_after))
thickness_change = thickness_after - thickness_before   #插层后厚度的变化量
#print('插层后厚度的变化量:', thickness_change, len(thickness_change))    

# plt.figure(figsize=(10, 6))
plt.plot(thickness_before, label='插层前')
plt.plot(thickness_after, label='插层后')
plt.ylabel('厚度')
plt.xlabel('组数')
plt.legend()
plt.show() 
thickness_before
```

```python
#孔隙率
porosity = pd.read_excel(io='./C题数据.xlsx', usecols=['孔隙率（%）'])
porosity_before = np.array(porosity)[:-1:2]
porosity_after = np.array(porosity)[1::2]
#print(porosity_before, len(porosity_before))
#print(porosity_after, len(porosity_after))
porosity_change = porosity_after - porosity_before   #插层后孔隙率的变化量

plt.plot(porosity_before, label='插层前')
plt.plot(porosity_after, label='插层后')
plt.ylabel('孔隙率')
plt.xlabel('组数')
plt.legend()
plt.show() 
```

```python
#压缩回弹性率
compression_resilience_ratio = pd.read_excel(io='./C题数据.xlsx', usecols=['压缩回弹性率（%）'])
compression_resilience_ratio_before = np.array(compression_resilience_ratio)[:-1:2]    #插层之前的压缩回弹性率
compression_resilience_ratio_after = np.array(compression_resilience_ratio)[1::2]      #插层之后的压缩回弹性率
compression_resilience_ratio_change = compression_resilience_ratio_after - compression_resilience_ratio_before #插层后压缩回弹性率的变化量

plt.plot(compression_resilience_ratio_before, label='插层前')
plt.plot(compression_resilience_ratio_after, label='插层后')
plt.ylabel('压缩回弹性率')
plt.xlabel('组数')
plt.legend()
plt.show() 
```

```python
#读取插层率
intercalation = pd.read_excel(io='./C题数据.xlsx', usecols=['插层率（%）'])
intercalation_rate = np.array(intercalation)[1::2]
print(intercalation_rate, len(intercalation_rate))
```

```python
#插层对厚度的影响
plt.figure(figsize=(10, 6))
# sorted(intercalation)
# tmp=pd.DataFrame()
# tmp[1]=intercalation_rate
# tmp[2]=thickness_change
# tmp.sort_values(by=1)
# plt.plot(tmp['1'], tmp['2'], c='red')
plt.scatter(intercalation_rate, thickness_change)
plt.xlabel('插层率')
plt.ylabel('厚度变化量')
plt.show()
```

```python
#插层对孔隙率的影响
tmp=pd.DataFrame()
tmp[['1']]=(intercalation_rate)
tmp[['2']]=porosity_change
tmp = tmp.sort_values(by='1')

plt.figure(figsize=(10, 6))
# plt.plot(tmp[['1']], tmp[['2']],c='red')
plt.scatter(intercalation_rate, porosity_change)
plt.xlabel('插层率')
plt.ylabel('孔隙率变化量')
plt.show()
```

```python
#插层对压缩回弹性率的影响
tmp=pd.DataFrame()
tmp[['1']]=(intercalation_rate)
tmp[['2']]=compression_resilience_ratio_change
tmp = tmp.sort_values(by='1')

plt.figure(figsize=(10, 6))
#plt.plot(tmp[['1']], tmp[['2']], c='red')
plt.scatter(intercalation_rate, compression_resilience_ratio_change)
plt.xlabel('插层率')
plt.ylabel('压缩回弹性率变化量')
plt.show()
```
